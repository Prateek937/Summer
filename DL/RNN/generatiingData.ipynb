```python
import numpy as np
```

```python
# Returns an array in range(num1, num2)
# Return evenly spaced values within a given interval.
time = np.arange(0, 1500)
```

```python
time
```

```python
# Creating a sinusoidal data
profit = np.sin(0.02*time)
```

```python
profit
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(profit)
```

```python
# Adding the noise to the data
# np.random.uniform() will return the reandom values between the low and high

profitWithNoise = np.sin(0.02*time) + np.random.uniform(low=-1, high=1, size=(1500,))
```

```python
profitWithNoise
```

```python
plt.plot(profit, color='green')
```

```python
plt.plot(profitWithNoise)
```
